```python

import os
from PIL import Image
from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt

# load in datasets from drive
# download dataset from kaggle and transfer the test and train folders to my drive
data_path_test_muffin = "/content/drive/MyDrive/test/muffin"
data_path_test_chihuahua = "/content/drive/MyDrive/test/chihuahua"
data_path_train_muffin = "/content/drive/MyDrive/train/muffin"
data_path_train_chihuahua = "/content/drive/MyDrive/train/chihuahua"

# assign the folders to variables
test_muffin_files = os.listdir(data_path_test_muffin)
test_chihuahua_files = os.listdir(data_path_test_chihuahua)
train_muffin_files = os.listdir(data_path_train_muffin)
train_chihuahua_files = os.listdir(data_path_train_chihuahua)


# counts the number of images in each folder
num_images_test_muffin = len(test_muffin_files)
num_images_test_chihuahua = len(test_chihuahua_files)
num_images_train_muffin = len(train_muffin_files)
num_images_train_chihuahua = len(train_chihuahua_files)

print("Number of classes: 2")

print(f"Number of images of muffins: {num_images_test_muffin + num_images_train_muffin}")
print(f"Number of images of chihuahuas: {num_images_test_chihuahua + num_images_train_chihuahua}")

print(f"Number of testing: {num_images_test_muffin + num_images_test_chihuahua}")
print(f"Number of training: {num_images_train_chihuahua + num_images_train_muffin}")
```

```python

# graph the number of muffin and chihuahua images
heights = [num_images_test_muffin + num_images_train_muffin, num_images_test_chihuahua + num_images_train_chihuahua]
plt.bar(['Muffin','Chihuahua'],heights)
plt.title("Number of Muffin and Chihuahua Images")
plt.ylabel("Counts")
plt.show()
```

```python
# shows an example of the muffin class
for filename in train_muffin_files:
    image_path = os.path.join(data_path_train_muffin, filename)
    image = Image.open(image_path)
    plt.imshow(image) # to see multiple images, add images to a list
    plt.title(filename)
    plt.show()
    break
```

```python
# show example of chihuahua class
for filename in train_chihuahua_files:
    image_path = os.path.join(data_path_train_chihuahua, filename)
    image = Image.open(image_path)
    plt.imshow(image)
    plt.title(filename)
    plt.show()
    break
```

```python
# add image size to lists
image_size_width = []
image_size_height = []

# loops through each folder and places the sizes in the corresponding list
for filename in test_muffin_files:
    image_path = os.path.join(data_path_test_muffin, filename)
    image = Image.open(image_path)
    width, height = image.size
    #print(f"Image: {filename}, Size: {width} x {height}")
    image_size_width.append(width)
    image_size_height.append(height)

for filename in test_chihuahua_files:
    image_path = os.path.join(data_path_test_chihuahua, filename)
    image = Image.open(image_path)
    width, height = image.size
    #print(f"Image: {filename}, Size: {width} x {height}")
    image_size_width.append(width)
    image_size_height.append(height)

for filename in train_chihuahua_files:
    image_path = os.path.join(data_path_train_chihuahua, filename)
    image = Image.open(image_path)
    width, height = image.size
    #print(f"Image: {filename}, Size: {width} x {height}")
    image_size_width.append(width)
    image_size_height.append(height)

for filename in train_muffin_files:
    image_path = os.path.join(data_path_train_muffin, filename)
    image = Image.open(image_path)
    width, height = image.size
    #print(f"Image: {filename}, Size: {width} x {height}")
    image_size_width.append(width)
    image_size_height.append(height)
```

```python
# Visualize the image sizes: from the plot we see that the image sizes are not equal and we need to convert all images to a 640x480
plt.scatter(image_size_width, image_size_height)
plt.title("Image Sizes")
plt.ylabel("Image Height")
plt.xlabel("Image Width")
```
